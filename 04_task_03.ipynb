```python
import numpy as np

n = 30
params = 7

sizes = np.random.randint(3, 50, size=n)
X_data = [np.random.randint(0, 10, size=(i, params)) for i in sizes]
y_data = [np.random.randint(10, 20, size=i) for i in sizes]

batch_size = 5
```

```python
class BatchGenerator(object):
    def __init__(self, X_data, y_data, batch_size, vector):
        self.X_data = X_data
        self.y_data = y_data
        self.batch_size = batch_size
        self.vector = vector
        
        # your code

    def generate(self):
        
        def func_list_invert(item):
            if len(item) <= self.vector:
                return []

            batch_size_list = list(range(self.vector-batch_size//2, self.vector+batch_size//2+1))
            batch_size_list = list(map(lambda x: 0 if x < 0 else x, batch_size_list))
            batch_size_list

            temp_list = list(range(0,len(item)))
            batch_size_list_state = list(map(lambda x: x in temp_list, batch_size_list))

            list_invert = []
            for item2 in list(range(0,len(batch_size_list))):
                if batch_size_list_state[item2] == False:
                    list_invert.append(batch_size_list[len(batch_size_list)-item2-1])
                else:
                    list_invert.append(batch_size_list[item2])
            return list_invert
        
        self.X_data = [[item[i] for i in func_list_invert(item)] for item in X_data]
        self.y_data = [[item[i] for i in func_list_invert(item)] for item in y_data]
        
        return self.X_data, self.y_data
#         your code
```

```python
batch_generator = BatchGenerator(X_data, y_data, batch_size=7, vector=4)
```
